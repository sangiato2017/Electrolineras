```python
import time
import os
import sys
import numpy as np
import pandas as pd
import networkx as nx
import osmnx as ox
import matplotlib as plt
import threading
```

```python
#============Funciones============#

#============Cargar el grafo ============#

G = None
cargando = False

def _cargar_grafo():
    global G, cargando
    cargando = True
    try:
        G = ox.load_graphml("red_vial_ambucaramanga/red_vial_amb.graphml")
    except ValueError:
        print("\n[-] Error cargando grafo")
    finally:
        cargando = False


def cargar_grafo_con_animacion():
    global cargando
    
    hilo = threading.Thread(target=_cargar_grafo)
    hilo.start()
    
    
    print("Cargando red vial", end="")
    while cargando:
        print(".", end="", flush=True)
        time.sleep(0.5)
    
    print("\n[+] ¡Grafo cargado exitosamente!")
    time.sleep(0.5)

#============Menú============#

def mostrar_menu(key):
    estado_grafo = "[OK]" if G is not None else "[cargando...]" if cargando else "[no cargado]"
    print("\n" + "="*40 + "\n")
    print("SISTEMA DE GESTIÓN DE ELECTROLINERAS")
    print("\n" + "="*40 + "\n" )
    print(f"  Red vial: {estado_grafo}")
    print("\n" + "="*40 + "\n" )
    print("1 |Cargar datos de la red vial de la ciudad")
    print("2 |Configurar parametros para la simulación de recorridos")
    print("3 |Ejecutar simulación de recorridos")
    print("4 |Mostrar estadisticas y resultados")
    print("5 |Salir")
    print("\n" + "="*40)

def simularProceso(seg, mns):
    print(f"\n{mns}", end = "")
    pts = 5
    timePts = seg / pts
    for i in range(pts):
        time.sleep(timePts)
        print(".", end = "")
        sys.stdout.flush()
    time.sleep(seg)
    print("\n\n[+] ¡Completado!")
    time.sleep(seg)

#============Validaciones============#

def errorSalida():
    seg = 1
    mns = "Regresando al menú principal"
    print(f"\n{mns}", end = "")
    pts = 3
    timePts = seg / pts
    for i in range(pts):
        time.sleep(timePts)
        print(".", end = "")
        sys.stdout.flush()

def enteroPosi(me):
    while True:
        ent = input(me)
        try:
            num = int(ent)
            if(num > 0):
                return num
            else:
                print("\n[-] Error: el numero debe ser mayor a 0.")
        except ValueError:
            print(f"\n[-] Error: {ent} no es un numero valido.")

def floatPosi(me):
    while True:
        ent = input(me)
        try:
            num = float(ent)
            if(num > 0 and num <= 100):
                return num
            else:
                print("\n[-] Error: el numero debe estar entre 0 y 1.")
        except ValueError:
            print(f"\n[-] Error: {ent} no es un numero valido.")

def salida(seg, mns):
    print(f"\n{mns}", end = "")
    pts = 5
    timePts = seg / pts
    for i in range(pts):
        time.sleep(timePts)
        print(".", end = "")
        sys.stdout.flush()
    time.sleep(seg)
```

```python
#============opciones del menú============#
OUTPUT_DIR = "red_vial_ambucaramanga"
os.makedirs(OUTPUT_DIR, exist_ok=True)

def opcion_1():
    global G, cargando
    
    cargar_grafo_con_animacion()


    
```

```python
#============Principal============#
def menu_principal():
    key = 0
    while True:
        mostrar_menu(key)
        op = enteroPosi("\nElija una opcion (1-5): ")
        match op:
            case 1:
                opcion_1()
                if G is not None and key == 0:
                    key = 1
                
            case 2:
                if(key == 1 or key == 2 or key == 3 or key == 4):
                    
                    n = enteroPosi("\nIngrese el numero de recorridos a simular: ")
                    
                    c = 0
                    
                    while (c != 2):
                        
                        print("\n" + "="*40 + "\n")
                        print("PUNTOS DE REFERENCIA")
                        print("\n" + "="*40 + "\n" )
                        
                        
                        c += 1
                    
                    
                    
                    simularProceso(1, "\nProcesando")
                    print("\n¡Configuración guradada con exito!")
                    time.sleep(0.5)
                    print(f"\nSe ejecutaran {n} recorridos ")
                    time.sleep(1)
                    
                    key = 2
                else:
                    print("\n[-] Error: No se han cargado datos" )
                    errorSalida()
            case 3:
                if (key == 2 or key == 3 or key == 4):
                    simularProceso(3, "\nEjecutando simulacion de recorridos")
                    
                    key = 3
                else:
                    print("\n[-] Error: No se han cargado datos o no se an configurado parametros" )
                    errorSalida()
            case 4:
                if(key == 3 or key == 4):
                    simularProceso(2, "\nGenerando estadisticas y resultados")
                else:
                    print("\n[-] Error: No se han simulado recorridos" )
                    errorSalida()
            case 5:
                salida(1, "\nSaliendo")
                break
            case _:
                print("\n[-] Error: Opcion invalida")
                errorSalida()
        
```

```python
#ejecutares

menu_principal()
```
